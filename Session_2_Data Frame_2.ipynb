```python
dtypes = visa_df.dtypes
keys = dtypes.keys()
values = dtypes.values
num,cat=[],[]
for key,value in zip(keys,values):
    if value == 'object':
        cat.append(key)
    else:
        num.append(key)
cat,num
```

```python
dtypes = visa_df.dtypes
keys = dtypes.keys()
values = dtypes.values
cat = [key for key,value in zip(keys,values) if value == 'object']
num = [key for key,value in zip(keys,values) if value != 'object']
```

```python
cat,num
```

# take

```python
# Select rows 100,200,300
visa_df.take([100,200,300])
```

```python
# Select continent, prevailing_wage, case_status
cols = [1,8,11]
visa_df.take(cols,axis=1)
```

```python
# Do both of the above in one program
visa_df.take([100,200,300])
cols = [1,8,11]
visa_df.take(cols,axis=1).take([100,200,300])
```

```python
# Do both of the above for a certain range of rows
rows = range(2,5)
cols = [1,8,11]
visa_df.take(cols,axis=1).take(rows)
```
