**Step-1: Import Packages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step-2: Create a Dataframe using List**

```python
names = ['Ramesh', 'Suresh','Mahesh']
age = [20,21,22]
city= ['Tirupati', 'Banglore', 'Hyderabad']
pd.DataFrame(zip(names,age,city))
```

**Step-3: Insert the Columns**

```python
import pandas as pd
names = ['Ramesh', 'Suresh','Mahesh']
age = [20,21,22]
city= ['Tirupati', 'Banglore', 'Hyderabad']
cols = ['Names','Age','City']
pd.DataFrame(zip(names,age,city), columns = cols)
```

**Step-4: Modify the Index**

```python
names = ['Ramesh', 'Suresh','Mahesh']
age = [20,21,22]
city= ['Tirupati', 'Banglore', 'Hyderabad']
cols = ['Names','Age','City']
idx = ['A','B','C']
pd.DataFrame(zip(names,age,city), columns = cols, index =idx)
```

```python
names = ['Ramesh','Suresh','Mahesh']
age = [25,35,45]
city = ['Tirupati','Chennai', 'Hyderabad']
cols = ['Name','Age','City']
idx = ['A','B','C']
df1=pd.DataFrame(zip(names,age,city),columns=cols, index=idx)
```

```python
type(df1)
```
