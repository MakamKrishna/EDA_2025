```python
import pandas as pd
d1 = {'ID' : [1,2,3],
     'Name': ['Chandu', 'Pawan', 'Bansal']}
d2 = {'ID' : [1,2,4],
     'Age': [20,25,30]}

df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on = 'ID', how = 'inner')
```

```python
pd.merge(df1,df2, on = 'ID', how = 'outer')
```

```python
pd.merge(df1,df2, on = 'ID', how = 'left')
```

```python
pd.merge(df1,df2, on = 'ID', how = 'right')
```
