```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
visa_df = pd.read_csv(r"C:\Users\Acer\OneDrive\Data Science\Data Analysis\Numpy\Visadataset.csv")
cat = visa_df.select_dtypes(include = 'object').columns
num = visa_df.select_dtypes(exclude = 'object').columns
```

# map method

```python
visa_df['case_status'].unique()
```

```python
dict1 = { 'Certified':0,'Denied':1}
dict1
```

```python
visa_df['case_status'] = visa_df['case_status'].map(dict1)
```

```python
visa_df
```

```python
labels = visa_df['continent'].unique()
d = {}
for i in range(len(labels)):
    d[labels[i]] = i
d
```

```python
labels = visa_df['continent'].unique()
count = visa_df['continent'].nunique()
d = {labels[i]:i for i in range(len(labels))}
visa_df['continent'] = visa_df['continent'].map(d)
visa_df
```

```python
for i in cat[1:]:
    labels = visa_df[i].unique()
    d = {labels[i]:i for i in range(len(labels))}
    visa_df[i] = visa_df[i].map(d)
visa_df
```

# Label Encoder

```python
# Step-1: import the package
from sklearn.preprocessing import LabelEncoder
# Step-2: Save the package
le = LabelEncoder()
# Step-3: fit and fit_transform
    # fit means finding some values 
    # transform means : transform apply those values
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cat[1:]:
    visa_df[i] = le.fit_transform(visa_df[i])
visa_df
```
